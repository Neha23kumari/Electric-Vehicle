# univariate analysis for categorical columns

```python
# Loop through each column in the DataFrame
for column in df.columns:
    # Check if the column's data type is object (categorical)
    if df[column].dtype == "object":
        # Create a bar plot for the top 10 unique values in the column
        value_counts = df[column].value_counts().nlargest(10)  # Get the top 10 values
        sns.barplot(x=value_counts.index, y=value_counts.values)
        
        # Set the title and format the plot
        plt.title(f"Distribution of {column}")
        plt.xticks(rotation=90)
        plt.xlabel(column)
        plt.ylabel('Frequency')
        
        # Display the plot
        plt.show()
```

# univariate analysis for numerical columns

```python

numerical_columns = ['Postal_Code', 'Model_Year', 'Electric_Range', 'Base_MSRP', 'DOL_Vehicle_ID', '2020_Census_Tract']

# Set up the figure and axes for a 3x2 grid of subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 10))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Loop through each numerical column and its corresponding subplot
for i, column in enumerate(numerical_columns):
    df[column].hist(bins=20, ax=axes[i])
    axes[i].set_title(f'Histogram of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frequency')

# Adjust the layout to prevent overlapping
plt.tight_layout()
plt.show()
```

# step 2-Create a Choropleth using plotly.express to display the number of EV vehicles based on location.


```python
import plotly.express as px

# Group by 'City' and 'State' to get the count of electric vehicles
city_ev_count = df.groupby(['City', 'State']).size().reset_index(name="Ev Count")

# Create a choropleth map using Plotly
fig = px.choropleth(
    city_ev_count,
    locations='State',
    locationmode='USA-states',
    color='Ev Count',
    hover_name='City',
    hover_data={'State': True, 'Ev Count': True},
    scope='usa',
    title="Number of Electric Vehicles by City and State",
    color_continuous_scale=px.colors.qualitative.Alphabet
)

# Show the figure
fig.show()
```

# Step 3-Create a Racing Bar Plot to display the animation of EV Make and its count each year.

```python
import plotly.express as px

# Group the data by 'Model year' and 'Make' and count occurrences
df_grouped = df.groupby(['Model_Year', 'Make']).size().reset_index(name='Count')

# Create an animated bar chart using Plotly
fig = px.bar(
    df_grouped,
    x='Count',
    y='Make',
    color='Make',
    animation_frame='Model_Year',
    range_x=[0, df_grouped['Count'].max() + 1],
    title='Count of EV Makes Over the Years',
    labels={'Count': 'Count of EVs', 'Make': 'EV Make'},
    text='Count'
)

# Display the figure
fig.show()
```
